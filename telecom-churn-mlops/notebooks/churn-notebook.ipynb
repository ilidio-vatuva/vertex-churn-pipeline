```python
from google.cloud import aiplatform

PROJECT_ID = "vertex-churn-pipeline"
REGION = "us-central1"

aiplatform.init(project=PROJECT_ID, location=REGION)
```

```python
fs = aiplatform.Featurestore.create(
    featurestore_id="telco_fs",
    online_store_fixed_node_count=1
)

customer = fs.create_entity_type(entity_type_id="customer")

customer.create_feature("tenure_months", "INT64")
customer.create_feature("monthly_charges", "DOUBLE")
customer.create_feature("total_data_gb", "DOUBLE")
customer.create_feature("support_tickets_30d", "INT64")
customer.create_feature("late_payments_6m", "INT64")
```

```python
customer.batch_import_feature_values(
    bigquery_source_uri=f"bq://{PROJECT_ID}.telco_ml.customer_history",
    entity_id_field="customer_id",
    feature_specs=[
        {"id": "tenure_months"},
        {"id": "monthly_charges"},
        {"id": "total_data_gb"},
        {"id": "support_tickets_30d"},
        {"id": "late_payments_6m"},
    ],
)
```
